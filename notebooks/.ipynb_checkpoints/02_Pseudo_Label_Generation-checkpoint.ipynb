```python
import pandas as pd
import numpy as np

from sentence_transformers import SentenceTransformer
from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/customer_support_tickets.csv")

df.head()
```

```python
df["Combined_Text"] = (
    df["Ticket_Subject"]
    + " "
    + df["Ticket_Description"]
)

df["Combined_Text"].head()
```

```python
severity_keywords = {
    "outage":5,
    "fraud":5,
    "hacked":5,
    "stolen":5,
    "payment failed":4,
    "unauthorized":4,
    "error":3,
    "failed":3,
    "urgent":2,
    "asap":1,
    "login":1,
    "password":1
}
```

```python
def keyword_score(text):
    text = text.lower()
    score = 0

    for word, value in severity_keywords.items():
        if word in text:
            score += value

    return score
```

```python
df["Keyword_Score"] = df["Combined_Text"].apply(keyword_score)

df["Keyword_Score"].describe()
```

```python
plt.figure(figsize=(8,4))

plt.hist(
    df["Keyword_Score"],
    bins=20,
    edgecolor="black"
)

plt.title("Keyword Severity Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.show()
```

```python
def resolution_score(hours):

    if hours <= 12:
        return 3

    elif hours <= 48:
        return 2

    else:
        return 1
```

```python
df["Resolution_Score"] = (
    df["Resolution_Time_Hours"]
    .apply(resolution_score)
)

df["Resolution_Score"].value_counts()
```

```python
df["Resolution_Score"].value_counts().sort_index().plot(
    kind="bar"
)

plt.title("Resolution Severity Score")
plt.show()
```

```python
model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)
```

```python
embeddings = model.encode(
    df["Combined_Text"].tolist(),
    show_progress_bar=True
)
```
